**4. Create a Bokeh histogram to visualize the distribution of the given data.**

```python
#import pandas as pd
#import numpy as np
#imort matplotlib as plt
#import plotly.graph objects as go
import plotly.express as px
!pip install numpy
import numpy as np

!pip install bokeh
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = bokeh.plotting.figure(title = "Histogram", x_axis_label = "Value", y_axis_label = "Frequency")
p.quad(top = hist, bottom = 0, left = edges[:-1], right = edges[1:], fill_color = "pink", line_color = "purple")
bokeh.plotting.show(p)
```

```python
#import pandas as pd
#import numpy as np
#imort matplotlib as plt
#import plotly.graph objects as go
import plotly.express as px
!pip install numpy
import numpy as np

!pip install bokeh
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

p = bokeh.plotting.figure(title = "Heatmap", x_axis_label = "X-axis", y_axis_label = "Y-axis")
p.image(image = [data_heatmap], x = [xx.min()], y = [yy.min()], dw = [xx.ptp()], dh = [yy.ptp()], palette = "Spectral11")
bokeh.plotting.show(p)
```
