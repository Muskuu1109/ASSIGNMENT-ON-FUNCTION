### 1. Difference Between `np.array`, `np.asarray`, and `np.asanyarray`

- **`np.array()`**:
  - **Purpose**: Creates a new NumPy array.
  - **Behavior**: Always copies the input data unless `copy=False` is explicitly specified. This means that if you pass a list or another array, it creates a new array with a new copy of the data.
  - **Usage**: Useful when you want a new array and do not want to affect the original data.
  - **Example**:
    ```python
    import numpy as np
    a = [1, 2, 3]
    b = np.array(a)
    ```

- **`np.asarray()`**:
  - **Purpose**: Converts the input to a NumPy array.
  - **Behavior**: Does not copy the data if the input is already an array or array-like object. It will only create a new array if necessary. If the input is already a NumPy array, it will return the same array.
  - **Usage**: Useful when you want to ensure that the input is a NumPy array but prefer to avoid unnecessary copying.
  - **Example**:
    ```python
    import numpy as np
    a = np.array([1, 2, 3])
    b = np.asarray(a)  # b is the same array as a, no copy is made
    ```

- **`np.asanyarray()`**:
  - **Purpose**: Converts the input to an array, but it allows for subclasses of `np.ndarray` to be preserved.
  - **Behavior**: Similar to `np.asarray()`, but it respects array subclasses. This means if you have a subclass of `np.ndarray`, `np.asanyarray()` will preserve that subclass, while `np.asarray()` will not.
  - **Usage**: Useful when working with array subclasses and you want to ensure that these subclasses are not stripped away.
  - **Example**:
    ```python
    import numpy as np
    class MyArray(np.ndarray):
        pass
    a = np.array([1, 2, 3]).view(MyArray)
    b = np.asanyarray(a)  # b is of type MyArray, same as a
    ```

### 2. Difference Between Deep Copy and Shallow Copy

- **Shallow Copy**:
  - **Definition**: Creates a new object, but does not create copies of nested objects (i.e., it only creates a new reference to the nested objects).
  - **Behavior**: Changes to the nested objects will be reflected in both the original and the shallow-copied objects.
  - **Usage**: Often used when you want a new container object but want to share the inner objects.
  - **Example**:
    ```python
    import copy

    original = [[1, 2, 3], [4, 5, 6]]
    shallow_copied = copy.copy(original)
    
    # Modify the nested list
    shallow_copied[0][0] = 99

    print("Original:", original)  # Output: Original: [[99, 2, 3], [4, 5, 6]]
    print("Shallow Copied:", shallow_copied)  # Output: Shallow Copied: [[99, 2, 3], [4, 5, 6]]
    ```

- **Deep Copy**:
  - **Definition**: Creates a new object and also recursively creates copies of all nested objects.
  - **Behavior**: Changes to the nested objects in the deep-copied object will not affect the original object.
  - **Usage**: Used when you need a completely independent copy of the original object, including all nested objects.
  

````python
#example

import copy

original = [[1, 2, 3], [4, 5, 6]]
deep_copied = copy.deepcopy(original)

# Modify the nested list
deep_copied[0][0] = 99

print("Original:", original)  # Output: Original: [[1, 2, 3], [4, 5, 6]]
print("Deep Copied:", deep_copied)  # Output: Deep Copied: [[99, 2, 3], [4, 5, 6]]
````

**17.Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot**

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)


data = pd.DataFrame({'x': x, 'y': y})

def determine_quadrant(row):
    if row['x'] > 0 and row['y'] > 0:
        return 'Quadrant I'
    elif row['x'] < 0 and row['y'] > 0:
        return 'Quadrant II'
    elif row['x'] < 0 and row['y'] < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

data['Quadrant'] = data.apply(determine_quadrant, axis=1)


plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='viridis', s=100)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrant')
plt.show()
````
